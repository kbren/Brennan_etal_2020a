```python
anom_start = 1979
anom_end = 2013
chosen_dset = ['GIS','BE','CRU']

# NH surface area in M km^2 from concentration in percentage
nharea = 2*np.pi*(6380**2)/1e8
```

### Load Reconstructions: 

```python
path = ('/home/disk/p/mkb22/nobackup/LMR_output/reanalysis_reconstruction_data/'+
        'annual/Brennan_etal_2019/full_reconstructions/')

inf_name_mpi = ['1_8']
inf_name_ccsm4 = ['2_6']
loc_mpi = [15000]
loc_ccsm4 = [15000]
iteration = ['a','b','c','d','e']

name_mpi = {}
name_ccsm4 = {}
name_dict_mpi = {}
name_dict_ccsm4 = {}
tot=0
           
# for ninf in range(len(inf_name)):
#     for nloc in range(len(loc)):
for it in iteration:
    name_mpi[tot] = ('mpi/sic_mpi_annual_recon_1850_2018_cru_gis_be_R0_4_10deg_inf'+
               inf_name_mpi[0]+'_loc'+str(loc_mpi[0])+it+'.pkl')
    name_ccsm4[tot] = ('ccsm4/sic_ccsm4_annual_recon_1850_2018_cru_gis_be_R0_4_10deg_inf'+
               inf_name_ccsm4[0]+'_loc'+str(loc_ccsm4[0])+it+'.pkl')

    name_dict_mpi[tot] = ('MPI R0.4 inf'+inf_name_mpi[0]+' loc'+str(loc_mpi[0])+' '+it)
    name_dict_ccsm4[tot] = ('CCSM4 R0.4 inf'+inf_name_ccsm4[0]+' loc'+str(loc_ccsm4[0])+' '+it)

    tot = tot+1
```

#### MPI

```python
sia_mpi_anom={}
sia_mpi_97_5={}
sia_mpi_2_5={}
sie_mpi_anom={}
sie_mpi_97_5={}
sie_mpi_2_5={}
recon_years_all_mpi={}
sie_mpi_full_ens = {}

#for d,dset in enumerate(chosen_dset):
for i in range(len(name_dict_mpi)):
    [sia_mpi_anom[name_dict_mpi[i]], 
     sia_mpi_97_5[name_dict_mpi[i]], 
     sia_mpi_2_5[name_dict_mpi[i]], 
     sie_mpi_anom[name_dict_mpi[i]], 
     sie_mpi_97_5[name_dict_mpi[i]], 
     sie_mpi_2_5[name_dict_mpi[i]], 
     recon_years_all_mpi[name_dict_mpi[i]],
     sie_mpi_full_ens[name_dict_mpi[i]]] = rrutils.load_recon_sie_95(path+name_mpi[i],chosen_dset,
                                                                     anom_start,anom_end,full=True)
```

#### CCSM4

```python
sia_ccsm4_anom={}
sia_ccsm4_97_5={}
sia_ccsm4_2_5={}
sie_ccsm4_anom={}
sie_ccsm4_97_5={}
sie_ccsm4_2_5={}
recon_years_all_ccsm4={}
sie_ccsm4_full_ens = {}

for i in range(len(name_dict_ccsm4)):
    [sia_ccsm4_anom[name_dict_ccsm4[i]], 
     sia_ccsm4_97_5[name_dict_ccsm4[i]], 
     sia_ccsm4_2_5[name_dict_ccsm4[i]], 
     sie_ccsm4_anom[name_dict_ccsm4[i]], 
     sie_ccsm4_97_5[name_dict_ccsm4[i]], 
     sie_ccsm4_2_5[name_dict_ccsm4[i]], 
     recon_years_all_ccsm4[name_dict_ccsm4[i]],
     sie_ccsm4_full_ens[name_dict_ccsm4[i]]] = rrutils.load_recon_sie_95(path+name_ccsm4[i],chosen_dset,
                                                                         anom_start,anom_end,full=True)
```

```python
sie_allit_anom = {}

dsize = {'GIS':136,'BE':164,'CRU':169}

for d,dset in enumerate(chosen_dset):
    sie_allit_fullens_ccsm4_anom = np.zeros((dsize[dset],200,5))
    sie_allit_fullens_mpi_anom = np.zeros((dsize[dset],200,5))

    for n in name_dict_mpi.keys():
        sie_mpi_anom_hold = np.zeros((dsize[dset],200))
        sie_ccsm4_anom_hold = np.zeros((dsize[dset],200))
    
        for j in range(sie_allit_fullens_mpi_anom.shape[1]):
            sie_mpi_anom_hold[:,j] = rrutils.find_anomalies(np.array(recon_years_all_mpi[name_dict_mpi[n]][dset]),
                                                                     sie_mpi_full_ens[name_dict_mpi[n]][dset][:,j],anom_start,anom_end)
            sie_ccsm4_anom_hold[:,j] = rrutils.find_anomalies(np.array(recon_years_all_ccsm4[name_dict_ccsm4[n]][dset]),
                                                                       sie_ccsm4_full_ens[name_dict_ccsm4[n]][dset][:,j],anom_start,anom_end)

        sie_allit_fullens_mpi_anom[:,:,n] = sie_mpi_anom_hold
        sie_allit_fullens_ccsm4_anom[:,:,n] = sie_ccsm4_anom_hold

    sie_allit_anom['CCSM4, '+dset] = sie_allit_fullens_ccsm4_anom
    sie_allit_anom['MPI, '+dset] = sie_allit_fullens_mpi_anom
```

```python
experiments = {}
experiments['ccsm4']=['CCSM4, GIS', 'CCSM4, BE', 'CCSM4, CRU']
experiments['mpi'] = ['MPI, GIS', 'MPI, BE', 'MPI, CRU']
```

```python
sie_97_5_full_anom = np.percentile(sie_allit_anom['MPI, CRU'],97.5,axis=1)
sie_2_5_full_anom = np.percentile(sie_allit_anom['MPI, CRU'],2.5,axis=1)

sie_97_5_full_ccsm4_anom = np.percentile(sie_allit_anom['CCSM4, CRU'],97.5,axis=1)
sie_2_5_full_ccsm4_anom = np.percentile(sie_allit_anom['CCSM4, CRU'],2.5,axis=1)
```

```python
sie_97_5_full_anom = {}
sie_2_5_full_anom = {}

for e,exp in enumerate(experiments['ccsm4']): 
    sie_anom_hold1 = np.reshape(sie_allit_anom[exp],(dsize[chosen_dset[e]],1000))
    sie_97_5_full_anom[exp] = np.percentile(sie_anom_hold1,97.5,axis=1)
    sie_2_5_full_anom[exp] = np.percentile(sie_anom_hold1,2.5,axis=1)

for e,exp in enumerate(experiments['mpi']): 
    sie_anom_hold2 = np.reshape(sie_allit_anom[exp],(dsize[chosen_dset[e]],1000))
    sie_97_5_full_anom[exp] = np.percentile(sie_anom_hold2,97.5,axis=1)
    sie_2_5_full_anom[exp] = np.percentile(sie_anom_hold2,2.5,axis=1)
```

### Plotting: 

```python
plt.figure(figsize=(16,10))
colors = ['orangered','coral','orange', 'blue','royalblue', 'deepskyblue']
tot=-1

for k in experiments.keys():
    for e,exp in enumerate(experiments[k]):
        tot = tot+1
        plt.plot(np.array(recon_years_all_mpi[name_dict_mpi[n]][chosen_dset[e]]),
                 np.nanmean(np.nanmean(sie_allit_anom[exp],axis=2),axis=1),label=exp, color=colors[tot])
        plt.fill_between(np.array(recon_years_all_mpi[name_dict_mpi[n]][chosen_dset[e]]),
                         sie_2_5_full_anom[exp],sie_97_5_full_anom[exp],alpha=0.4,color=colors[tot])
        plt.legend()
```
