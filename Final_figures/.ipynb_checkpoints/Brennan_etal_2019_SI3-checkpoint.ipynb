### MPI: 

```python
path = ('/home/disk/p/mkb22/nobackup/LMR_output/reanalysis_reconstruction_data/'+
        'annual/Brennan_etal_2019/full_reconstructions/')

inf_name_mpi = ['1_8']
inf_name_ccsm4 = ['2_6']
loc_mpi = [15000]
loc_ccsm4 = [15000]
iteration = ['a','b','c','d','e']

name_mpi = {}
name_ccsm4 = {}
name_dict_mpi = {}
name_dict_ccsm4 = {}
tot=0
           
# for ninf in range(len(inf_name)):
#     for nloc in range(len(loc)):
for it in iteration:
    name_mpi[tot] = ('mpi/sic_mpi_annual_recon_1850_2018_cru_gis_be_R0_4_10deg_inf'+
               inf_name_mpi[0]+'_loc'+str(loc_mpi[0])+it+'.pkl')
    name_ccsm4[tot] = ('ccsm4/sic_ccsm4_annual_recon_1850_2018_cru_gis_be_R0_4_10deg_inf'+
               inf_name_ccsm4[0]+'_loc'+str(loc_ccsm4[0])+it+'.pkl')

    name_dict_mpi[tot] = ('MPI R0.4 inf'+inf_name_mpi[0]+' loc'+str(loc_mpi[0])+' '+it)
    name_dict_ccsm4[tot] = ('CCSM4 R0.4 inf'+inf_name_ccsm4[0]+' loc'+str(loc_ccsm4[0])+' '+it)

    tot = tot+1
```

```python
sia_mpi_anom={}
sia_mpi_97_5={}
sia_mpi_2_5={}
sie_mpi_anom={}
sie_mpi_97_5={}
sie_mpi_2_5={}
recon_years_all_mpi={}
sie_mpi_full_ens = {}

#for d,dset in enumerate(chosen_dset):
for i in range(len(name_dict_mpi)):
    [sia_mpi_anom[name_dict_mpi[i]], 
     sia_mpi_97_5[name_dict_mpi[i]], 
     sia_mpi_2_5[name_dict_mpi[i]], 
     sie_mpi_anom[name_dict_mpi[i]], 
     sie_mpi_97_5[name_dict_mpi[i]], 
     sie_mpi_2_5[name_dict_mpi[i]], 
     recon_years_all_mpi[name_dict_mpi[i]],
     sie_mpi_full_ens[name_dict_mpi[i]]] = rrutils.load_recon_sie_95(path+name_mpi[i],chosen_dset,
                                                                     anom_start,anom_end,full=True)
```

```python
sia_ccsm4_anom={}
sia_ccsm4_97_5={}
sia_ccsm4_2_5={}
sie_ccsm4_anom={}
sie_ccsm4_97_5={}
sie_ccsm4_2_5={}
recon_years_all_ccsm4={}
sie_ccsm4_full_ens = {}

for i in range(len(name_dict_ccsm4)):
    [sia_ccsm4_anom[name_dict_ccsm4[i]], 
     sia_ccsm4_97_5[name_dict_ccsm4[i]], 
     sia_ccsm4_2_5[name_dict_ccsm4[i]], 
     sie_ccsm4_anom[name_dict_ccsm4[i]], 
     sie_ccsm4_97_5[name_dict_ccsm4[i]], 
     sie_ccsm4_2_5[name_dict_ccsm4[i]], 
     recon_years_all_ccsm4[name_dict_ccsm4[i]],
     sie_ccsm4_full_ens[name_dict_ccsm4[i]]] = rrutils.load_recon_sie_95(path+name_ccsm4[i],chosen_dset,
                                                                         anom_start,anom_end,full=True)
```

```python
sie_allit_anom = {}

for d,dset in enumerate(chosen_dset):
    for i in range(len(iteration)):
        if i is 0: 
            sie_allit_fullens_mpi = sie_mpi_full_ens[name_dict_mpi[0]][dset]
            sie_allit_fullens_ccsm4 = sie_ccsm4_full_ens[name_dict_ccsm4[0]][dset]
        else:
            sie_allit_fullens_mpi = np.concatenate((sie_allit_fullens_mpi,
                                                   sie_mpi_full_ens[name_dict_mpi[i]][dset]),axis=1)
            sie_allit_fullens_ccsm4 = np.concatenate((sie_allit_fullens_ccsm4,
                                                      sie_ccsm4_full_ens[name_dict_ccsm4[i]][dset]),axis=1)

    sie_allit_fullens_mpi_anom = np.zeros(sie_allit_fullens_mpi.shape)
    sie_allit_fullens_ccsm4_anom = np.zeros(sie_allit_fullens_ccsm4.shape)
    
    for i in range(sie_allit_fullens_mpi.shape[1]):
        sie_allit_fullens_mpi_anom[:,i] = rrutils.find_anomalies(np.array(recon_years_all_mpi[name_dict_mpi[0]][dset]),
                                                             sie_allit_fullens_mpi[:,i],anom_start,anom_end)
        sie_allit_fullens_ccsm4_anom[:,i] = rrutils.find_anomalies(np.array(recon_years_all_ccsm4[name_dict_ccsm4[0]][dset]),
                                                         sie_allit_fullens_ccsm4[:,i],anom_start,anom_end)
    
    sie_allit_anom['CCSM4, '+dset] = sie_allit_fullens_ccsm4_anom
    sie_allit_anom['MPI, '+dset] = sie_allit_fullens_mpi_anom
```

```python
experiments = {}
experiments['ccsm4']=['CCSM4, GIS', 'CCSM4, BE', 'CCSM4, CRU']
experiments['mpi'] = ['MPI, GIS', 'MPI, BE', 'MPI, CRU']
```

```python
sie_97_5_full_anom = np.percentile(sie_allit_anom['MPI, CRU'],97.5,axis=1)
sie_2_5_full_anom = np.percentile(sie_allit_anom['MPI, CRU'],2.5,axis=1)

sie_97_5_full_ccsm4_anom = np.percentile(sie_allit_anom['CCSM4, CRU'],97.5,axis=1)
sie_2_5_full_ccsm4_anom = np.percentile(sie_allit_anom['CCSM4, CRU'],2.5,axis=1)
```

### Calculate statistics: 

```python
stat_start = [1880,1850,1850]
stat_end = [2015,2013,2017]

ce_cru_sie = {}
corr_cru_sie = {}
var_cru_sie = {}

ce_cru_sie_dt = {}
corr_cru_sie_dt = {}
var_cru_sie_dt = {}

slope_cru_sie = {}
sat_slope = {}

priors = ['ccsm4','mpi']
n = name_dict_mpi[0]


for p in priors:
    for e,exp in enumerate(experiments[p]):
        var = np.nanmean(sie_allit_anom[exp],axis=1)
        [ce_cru_sie[exp], 
         corr_cru_sie[exp],
         var_cru_sie['sat'], 
         var_cru_sie[exp]] = rrutils.find_ce_corr(var, 
                                                  fet_sie_anom,#[chosen_dset[e]],
                                                  fet_time,
                                                  np.array(recon_years_all_mpi[n][chosen_dset[e]]),
                                                  anom_start,stat_end[e])

        [ce_cru_sie_dt[exp], 
         corr_cru_sie_dt[exp],
         var_cru_sie_dt['sat'], 
         var_cru_sie_dt[exp]] = rrutils.find_ce_corr(var, 
                                                     fet_sie_anom,
                                                     fet_time,
                                                     np.array(recon_years_all_mpi[n][chosen_dset[e]]),
                                                     anom_start,stat_end[e],detrend=True)

        slope_cru_sie_sat = []
        slope_cru_sie_etcw = []
        for i in range(sie_allit_anom[exp].shape[1]):
            slope_cru_sie_oneit = rrutils.find_trend(sie_allit_anom[exp][:,i],
                                                     np.array(recon_years_all_mpi[n][chosen_dset[e]]),
                                                     anom_start,stat_end[e])

            slope_cru_sie_oneit_etcw = rrutils.find_trend(sie_allit_anom[exp][:,i],
                                                          np.array(recon_years_all_mpi[n][chosen_dset[e]]),
                                                          1912,1937)

            slope_cru_sie_sat = np.append(slope_cru_sie_sat,slope_cru_sie_oneit)
            slope_cru_sie_etcw = np.append(slope_cru_sie_etcw,slope_cru_sie_oneit_etcw)
            
        a = {}
        a['sat era'] = slope_cru_sie_sat    
        slope_cru_sie[exp] = a
        slope_cru_sie[exp]['etcw'] = slope_cru_sie_etcw

        sat_slope[chosen_dset[e]] = rrutils.find_trend(fet_sie, fet_time, anom_start,stat_end[e])
walsh_etcw_slope = rrutils.find_trend(walsh_sie_anom, walsh_time, 1912,1937)
```

#### Calculate statistics between experiments that use the same prior but different observational records: 

```python
#recon_years_all_mpi
```

```python
# Calculate statistics for experiments with same prior but different observations assimilated 
cross_names = {}
cross_names[priors[1]] = ['MPI: cru/gis','MPI: cru/be','MPI: be/gis']
cross_names[priors[0]] = ['CCSM4: cru/gis','CCSM4: cru/be','CCSM4: be/gis']

ce_cross = {}
corr_cross = {}
var_cross = {}
order = {0:2,1:2,2:1}

for p in priors: 
    for e,exp in enumerate(experiments[p]):
        if e<2: eind = e
        else: eind = 0   
            
        var1 = np.nanmean(sie_allit_anom[experiments[p][order[e]]],axis=1)
        var2 = np.nanmean(sie_allit_anom[experiments[p][eind]],axis=1)
        
        astart = max(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[eind]][0],
                     recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[order[e]]][0])
        aend = min(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[eind]][-1],
                   recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[order[e]]][-1])
        
        [ce_cross[cross_names[p][e]], 
         corr_cross[cross_names[p][e]],
         _, 
         var_cross[cross_names[p][e]]] = rrutils.find_ce_corr(var1, var2,
                                             np.array(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[eind]]),
                                             np.array(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[order[e]]]),
                                             astart,aend)
```

```python
print('CE: ')
for p in priors:
    for n in (cross_names[p]):
        print(n + ' = '+ str(np.round(ce_cross[n],3)))
print('\n')    
print('R^2 values: ') 
for p in priors:
    for n in (cross_names[p]):
        print(n + ' = '+ str(np.round(corr_cross[n]**2,3)))
print('\n') 
print('Variance: ')
for p in priors:
    for n in (cross_names[p]):
        print(n + ' = '+ str(np.round(var_cross[n],3)))
```

```python
# Calculate statistics for experiments with same prior but different observations assimilated 
cross_names_tdata = {}
cross_names_tdata = ['CRU: mpi/ccsm4','GIS: mpi/ccsm4','BE: mpi/ccsm4']

for e,exp in enumerate(experiments[p]):  

    var1 = np.nanmean(sie_allit_anom[experiments['ccsm4'][e]],axis=1)
    var2 = np.nanmean(sie_allit_anom[experiments['mpi'][e]],axis=1)

    astart = recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[e]][0]
    aend = recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[eind]][-1]

    [ce_cross[cross_names_tdata[e]], 
     corr_cross[cross_names_tdata[e]],
     _, 
     var_cross[cross_names_tdata[e]]] = rrutils.find_ce_corr(var1, var2,
                                         np.array(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[e]]),
                                         np.array(recon_years_all_mpi[name_dict_mpi[0]][chosen_dset[e]]),
                                         astart,aend)
```

```python
print('CE: ')
for n in (cross_names_tdata):
    print(n + ' = '+ str(np.round(ce_cross[n],3)))
print('\n')    
print('R^2 values: ') 
for n in (cross_names_tdata):
    print(n + ' = '+ str(np.round(corr_cross[n]**2,3)))
print('\n') 
print('Variance: ')
for n in (cross_names_tdata):
    print(n + ' = '+ str(np.round(var_cross[n],3)))
```
