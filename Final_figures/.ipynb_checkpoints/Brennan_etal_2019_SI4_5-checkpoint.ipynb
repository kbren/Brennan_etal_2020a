```python
def inf_v_loc_plot(NXTICKS,XNAMES,NYTICKS,YNAMES,VAR,TITLE,LABEL,VMIN,VMAX,CMAP_R=False,sat=False):
 #   plt.figure(figsize=(6,4))
    
    if CMAP_R is False: 
        cmapr = 'PuBuGn'
    else:
        cmapr = 'PuBuGn_r'
        
    x_axis = np.arange(0,2*NXTICKS+2,2)
    y_axis = np.flip(np.arange(0,2*NYTICKS+2,2))
    
    xticks = np.arange(1,2*NXTICKS+1,2)
    yticks = np.flip(np.arange(1,2*NYTICKS+1,2))
    
    if sat is True:
        fig= plt.figure()
        ax = fig.add_subplot(1, 1, 1)
        #ax = plt.axes()
        ax.add_patch(plt.Rectangle((14, 0), 2, 12, fill=False, edgecolor='k', lw=2))
    
    plt.pcolormesh(x_axis,y_axis,VAR,cmap=cmapr,vmin=VMIN,vmax=VMAX)#,snap=snap_test)#, edgecolors='k')
    
    tot=-1
    for x in range(len(xnames)-1):
        for y in range(len(ynames)):
            tot = tot+1
            plt.text(x*2+1, loc_x[y], np.round(VAR,3)[y][x],ha='center',va='center', fontsize=14, color='w')
    
    if VAR.shape[1]>5:
        plt.text(11, 6, np.round(VAR[0,5],3),ha='center',va='center', fontsize=14, color='w')

    plt.yticks(yticks,YNAMES,fontsize=14)
    plt.xticks(xticks,XNAMES,fontsize=14)
    plt.xlabel('Inflation Factor', fontsize=14)
    plt.ylabel('Localization Length Scale (km)', fontsize=14)

    cbar = plt.colorbar().set_label(label=LABEL,size=14)
    plt.title(TITLE, fontsize=18)
```

```python
ce_cru_ccsm4_allit = {}
corr_cru_ccsm4_allit = {}
var_cru_ccsm4_allit = {}
slope_cru_ccsm4_allit = {}
intercept_cru_ccsm4_allit = {}

var_cru_dt_ccsm4_allit = {}

for n in name_iteration_ccsm4:
    name = name_iteration_ccsm4[n]
    shape = sie_ens_full_allit_ccsm4[name].shape
    sie_all_ccsm4 = np.reshape(sie_ens_full_allit_ccsm4[name],(shape[0],shape[1]*shape[2]))
    
    slope_cru_sie_sat = []
    ce_cru_sie_ccsm4 = []
    corr_cru_sie_ccsm4 = []
    var_cru_sie_ccsm4 = []
    var_cru_dt_sie_ccsm4 = []

    for it in range(shape[1]*shape[2]):
        sie_all_anom = rrutils.find_anomalies(np.arange(1979,2019,1), 
                                              sie_all_ccsm4[:,it],1979,2017)
        [ce_cru_ccsm4, corr_cru_ccsm4,
         var_cru_ccsm4_allit['sat'], 
         var_cru_ccsm4] = rrutils.find_ce_corr(sie_all_anom, fet_sie_anom,
                                             fet_time,np.arange(1979,2019,1),
                                             1979,2017) 
        
        [_, _,var_cru_dt_ccsm4_allit['sat'], 
         var_cru_dt_ccsm4] = rrutils.find_ce_corr(sie_all_anom,fet_sie_anom,
                                                  fet_time,np.arange(1979,2019,1), 
                                                  1979,2017,detrend=True)

        slope_cru_ccsm4 = rrutils.find_trend(sie_all_anom, np.arange(1979,2019,1),
                                             1979,2017, all_output=False)
        
        ce_cru_sie_ccsm4 = np.append(ce_cru_sie_ccsm4,ce_cru_ccsm4)  
        corr_cru_sie_ccsm4 = np.append(corr_cru_sie_ccsm4,corr_cru_ccsm4)  
        var_cru_sie_ccsm4 = np.append(var_cru_sie_ccsm4,var_cru_ccsm4)
        var_cru_dt_sie_ccsm4 = np.append(var_cru_dt_sie_ccsm4,var_cru_dt_ccsm4)
        slope_cru_sie_sat = np.append(slope_cru_sie_sat,slope_cru_ccsm4)    
    
    ce_cru_ccsm4_allit[name] = ce_cru_sie_ccsm4
    corr_cru_ccsm4_allit[name] = corr_cru_sie_ccsm4
    var_cru_ccsm4_allit[name] = var_cru_sie_ccsm4
    var_cru_dt_ccsm4_allit[name] = var_cru_dt_sie_ccsm4
    slope_cru_ccsm4_allit[name] = slope_cru_sie_sat
    
sat_slope = rrutils.find_trend(fet_sie_anom, fet_time, 1979,2017)
```
