```python
def scatter_fig(walsh_temp, walsh_si, walsh_time, sat_temp, sat_si, timebreak, ylabel, title, temp_dset, num):   
#    plt.figure(figsize=(7,6))
    if timebreak is False:
        if walsh_si.shape[0]<walsh_temp.shape[0]:
            walsh_temp = walsh_temp[:-1]
        else: 
            walsh_temp = walsh_temp 
            
        plt.scatter(walsh_temp,walsh_si,
                label='Walsh and '+temp_dset,
                color='k',alpha=0.5)
        plt.scatter(sat_temp,sat_si,
                    label='Satellite and '+temp_dset+' (1979-2017)',
                    color='crimson',alpha=0.6)
        
        [slope_obs, intercept_obs, 
         r_value_obs,_,_] = stats.linregress(sat_temp,sat_si)

        [slope_walsh, intercept_walsh, 
         r_value_walsh,_,_] = stats.linregress(walsh_temp,walsh_si)
        
        plt.plot(np.arange(-4,2.2,0.1),slope_walsh*np.arange(-4,2.2,0.1)+intercept_walsh,
                 color='k', linewidth=3)
        plt.plot(np.arange(-4,2.2,0.1),slope_obs*np.arange(-4,2.2,0.1)+intercept_obs,
                 color='crimson', linewidth=3)


        trend_sat = ('Satellite trend = ' + '{:,.3f}'.format(slope_obs)+
                     ',   $R^{2}$ = ' + '{:,.2f}'.format(r_value_obs**2))
        trend_walsh = ('Walsh trend     = ' + '{:,.3f}'.format(slope_walsh)+
                       ',   $R^{2}$ = ' + '{:,.2f}'.format(r_value_walsh**2))

        plt.gcf().text(0.42*num+0.14, 0.18, trend_sat, fontsize=12,color='crimson')
        plt.gcf().text(0.42*num+0.14, 0.14, trend_walsh, fontsize=12,color='k')
        
    else: 
        if temp_dset is 'GISTEMP':
            start = '1880'
        else:
            start = '1850'
        plt.scatter(walsh_temp[np.where(walsh_time <=timebreak)],
                    walsh_si[np.where(walsh_time <=timebreak)],
                    label='Walsh and '+temp_dset+' ('+start+'-'+str(timebreak)+')',
                    color='dimgray',#color='dimgray',
                    alpha=0.3)
        plt.scatter(walsh_temp[np.where(walsh_time >timebreak)] ,
                    walsh_si[np.where(walsh_time >timebreak)],
                    label='Walsh and '+temp_dset+' ('+str(timebreak+1)+'-2013)',
                    color='k',alpha=0.9)
        plt.scatter(sat_temp,sat_si,
                    label='Satellite and '+temp_dset+' (1979-2017)',
                    color='crimson',alpha=0.8)
    
        [slope_obs, intercept_obs, 
         r_value_obs,_,_] = stats.linregress(sat_temp,sat_si)

        [slope_walsh_early, intercept_walsh_early, 
         r_value_walsh_early,_,_] = stats.linregress(walsh_temp[np.where(walsh_time <=timebreak)],
                                                     walsh_si[np.where(walsh_time <=timebreak)])

        [slope_walsh_late, intercept_walsh_late, 
         r_value_walsh_late,_,_] = stats.linregress(walsh_temp[np.where(walsh_time >timebreak)],
                                                    walsh_si[np.where(walsh_time >timebreak)])
    
        plt.plot(np.arange(-4,2.2,0.1),slope_walsh_early*np.arange(-4,2.2,0.1)+intercept_walsh_early,
                 color='dimgray', linewidth=3)
        plt.plot(np.arange(-4,2.2,0.1),slope_walsh_late*np.arange(-4,2.2,0.1)+intercept_walsh_late,
                 color='k', linewidth=3)
        plt.plot(np.arange(-4,2.2,0.1),slope_obs*np.arange(-4,2.2,0.1)+intercept_obs,
                 color='crimson', linewidth=3)


        trend_sat = ('Satellite trend                   = ' + '{:,.3f}'.format(slope_obs)+
                     ',   $R^{2}$ = ' + '{:,.2f}'.format(r_value_obs**2))
        trend_walsh_early = ('Walsh trend ('+start+'-'+str(timebreak)+')  = ' + '{:,.3f}'.format(slope_walsh_early)+
                       ',   $R^{2}$ = ' + '{:,.2f}'.format(r_value_walsh_early**2))
        trend_walsh_late = ('Walsh trend ('+str(timebreak+1)+'-2013)  = ' + '{:,.3f}'.format(slope_walsh_late)+
                       ',   $R^{2}$ = ' + '{:,.2f}'.format(r_value_walsh_late**2))

        plt.gcf().text(0.48*num+0.07, 0.22, trend_sat, fontsize=12,color='crimson')
        plt.gcf().text(0.48*num+0.07, 0.18, trend_walsh_early, fontsize=12,color='dimgray')
        plt.gcf().text(0.48*num+0.07, 0.14, trend_walsh_late, fontsize=12,color='k')

#     plt.ylabel(ylabel, fontsize=14)
#     plt.xlabel('Arctic temperature anomalies (K)', fontsize=14)
#     plt.title(title, fontsize=18)
# #     plt.legend(fontsize=12,bbox_to_anchor=(1.73, 1))
#    plt.ylim(-3,4.5)
    # plt.xlim(-4,2.5)
```

### Winter slope significance: 

```python
import scipy as spy
```

```python
[slope_obs, intercept_obs,r_value_obs,_,_] = stats.linregress(cru_tas_winter_sat,fet_sie_winter_anom)

[slope_walsh, 
 intercept_walsh,
 r_value_walsh,p,_] = stats.linregress(cru_tas_winter_walsh[np.where(walsh_time[:-1] >1952)],
                                       walsh_sie_winter_anom[np.where(walsh_time[:-1] >1952)])

sat_era_pred = cru_tas_winter_sat*slope_obs+intercept_obs
walsh_era_pred = cru_tas_winter_walsh[np.where(walsh_time[:-1] >1952)]*slope_walsh+intercept_walsh

MSE_sat_era = np.square(np.subtract(fet_sie_winter_anom,sat_era_pred)).mean()
MSE_walsh_era = np.square(np.subtract(walsh_sie_winter_anom[np.where(walsh_time[:-1] >1952)],walsh_era_pred)).mean()

F = MSE_sat_era/MSE_walsh_era

p_value = spy.stats.f.cdf(F, cru_tas_winter_sat.shape[0], cru_tas_winter_walsh.shape[0])
p_value
```

```python
[slope_obs, intercept_obs,r_value_obs,_,_] = stats.linregress(cru_tas_winter_sat,fet_sie_winter_anom)

[slope_walsh, 
 intercept_walsh,
 r_value_walsh,p,_] = stats.linregress(cru_tas_winter_walsh[np.where(walsh_time[:-1] >1952)],
                                       walsh_sie_winter_anom[np.where(walsh_time[:-1] >1952)])

sat_era_pred = cru_tas_winter_sat*slope_obs+intercept_obs
walsh_era_pred = cru_tas_winter_walsh[np.where(walsh_time[:-1] >1952)]*slope_walsh+intercept_walsh

MSE_sat_era = (np.square(np.subtract(fet_sie_winter_anom,sat_era_pred)).mean())
MSE_walsh_era = (np.square(np.subtract(walsh_sie_winter_anom[np.where(walsh_time[:-1] >1952)],
                                       walsh_era_pred)).mean())

F = ((MSE_sat_era-MSE_walsh_era)/(walsh_era_pred.shape[0]-sat_era_pred.shape[0]))/(MSE_walsh_era/sat_era_pred.shape[0])

p_value = spy.stats.f.cdf(F, (walsh_era_pred.shape[0]-sat_era_pred.shape[0]), sat_era_pred.shape[0])
p_value
```
