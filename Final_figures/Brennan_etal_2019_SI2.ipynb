```python
###### 5 year running mean:
rn_time = 3

mn5_arc_cru_anom = np.zeros(mn_arc_cru_anom.shape[0]-rn_time)
mn5_arc_era_anom = np.zeros(mn_arc_era_anom.shape[0]-rn_time)
mn5_arc_noaa_anom = np.zeros(mn_arc_noaa_anom.shape[0]-rn_time)

for i in range(mn_arc_cru_anom.shape[0]-rn_time):
    mn5_arc_cru_anom[i] = np.nanmean(mn_arc_cru_anom[i:i+rn_time])
    
for i in range(mn_arc_era_anom.shape[0]-rn_time):
    mn5_arc_era_anom[i] = np.nanmean(mn_arc_era_anom[i:i+rn_time])
    
for i in range(mn_arc_noaa_anom.shape[0]-rn_time):
    mn5_arc_noaa_anom[i] = np.nanmean(mn_arc_noaa_anom[i:i+rn_time])
```
